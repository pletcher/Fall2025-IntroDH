```python
import json

with open("./data/my_network.json") as f:
    my_network = json.load(f)
```

```python
child_lists = []

def collect_child_lists(node):
    children = node.get("children")
    if children is not None:
        child_lists += children

        for child in children:
            collect_child_lists(child)

children = []
seen = set()

for l in child_lists:
    for node in l:
        if node["title"] in seen:
            continue
        else:
            seen.add(node["title"])

            children.append(node)
```

```python
from collections import Counter

parsed_articles = {}

for node in my_network:
    doc = nlp(node["content"])
    counts = Counter([t.pos_ for t in doc])
    node["number_of_pronouns"] = counts["PRON"]
    node["total_tokens"] = len(doc)
    parsed_articles[node["title"]] = node
```

```python
for title, article in parsed_articles.items():
    print(title)
    print(article["number_of_pronouns"], article["total_tokens"])
    print(f"% pronouns: {article["number_of_pronouns"] / article["total_tokens"]}")
```
